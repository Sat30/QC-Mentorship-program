## Input Setup

```python
import json

result = {}
with open('RESULT.json') as f:
    result = json.load(f)
    
QUBIT_SIZES = result['QUBIT_SIZES']
RUNTIMES = result['RUNTIMES']
MAX_QUBITS = result['MAX_QUBITS']
ERRORS = result['ERRORS']
```

## Sub Task 1


```python
import matplotlib.pyplot as plt

plt.figure(facecolor='#f0f0f0')  
plt.plot(QUBIT_SIZES[0], RUNTIMES[0], marker='s')
plt.title("Quantum Circuit Runtime Analysis")
plt.xlabel("Number of Qubits")
plt.ylabel("Average Runtime (seconds)")
plt.show()
```

## Sub Task 2

```python

plt.figure(facecolor='#f0f0f0')  
plt.plot(QUBIT_SIZES[1], RUNTIMES[1], marker='s')
plt.title("Quantum Circuit Runtime Analysis")
plt.xlabel("Number of Qubits")
plt.ylabel("Average Runtime (seconds)")
plt.show()
```

## Comparison Between Subtask-1 and Subtask-2

```python

plt.figure(facecolor='#f0f0f0')  
plt.plot(QUBIT_SIZES[0], RUNTIMES[0], marker='s', label='Sub Task 1')
plt.plot(QUBIT_SIZES[1], RUNTIMES[1], marker='s', label='Sub Task 2')
plt.title("Quantum Circuit Runtime Analysis")
plt.xlabel("Number of Qubits")
plt.ylabel("Average Runtime (seconds)")
plt.legend()
plt.show()
```

## Maximum Simulated Qubits

```python
approaches = ["Sub Task1", "Sub Task2"]

plt.bar(approaches, MAX_QUBITS, color=['blue', 'green'])
plt.title("Maximum Qubits Simulation Comparison")
plt.xlabel("Approach")
plt.ylabel("Maximum Qubits Simulated")

plt.show()
```

## Errors

```python
print("-"*100, "\n")
print(f"Simulation Error on Sub Task 1 for {MAX_QUBITS[0] + 1} Qubit is\n-- {ERRORS[0]}\n")
print("-"*100, "\n")
print(f"Simulation Error on Sub Task 2 for {MAX_QUBITS[1] + 1} Qubit is\n-- {ERRORS[1]}\n")
print("-"*100, "\n")
```
